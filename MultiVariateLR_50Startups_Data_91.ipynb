```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression as lr
from sklearn.model_selection import train_test_split as tts
```

```python
sd = pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
print(sd)
```

```python
sd['State'] = sd['State'].replace('New York', 0)
sd['State'] = sd['State'].replace('California', 1)
sd['State'] = sd['State'].replace('Florida', 2)
```

```python
#It can be observed that R&D Spend, Marketing Spend and Administration have higher correlated values than the rest columns
mlr = lr()
```

```python
X_train, X_test, y_train, y_test = tts(sd[['R&D Spend', 'Marketing Spend', 'Administration']], sd['Profit'], test_size = 0.3, random_state = 42)
```

```python
mlr.fit(X_train, y_train)
```

```python
y_train_pred = mlr.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score as r2
```

```python
print("Training Accuracies : ")
print("Mean Absolute Error : ", mae(y_train_pred, y_train))
print("Mean Squared Error : ", mse(y_train_pred, y_train))
print("Root Mean Squared Error : ", mse(y_train_pred, y_train)**0.5)
print("R2 Score : ", r2(y_train_pred, y_train))
```

```python
y_test_pred = mlr.predict(X_test)
```

```python
print("Testing Accuracies : ")
print("Mean Absolute Error : ", mae(y_test_pred, y_test))
print("Mean Squared Error : ", mse(y_test_pred, y_test))
print("Root Mean Squared Error : ", mse(y_test_pred, y_test)**0.5)
print("R2 Score : ", r2(y_test_pred, y_test))
```
